```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# Connection parameters
username = "root"
password = "Alexx123."
host = "localhost"
port = 3306  # must be integer
database = "restaurant_health_inspection"

# Encode password (safe for special characters)
password_encoded = quote_plus(password)

# Create SQLAlchemy engine
engine = create_engine(
    f"mysql+pymysql://{username}:{password_encoded}@{host}:{port}/{database}",
    connect_args={"local_infile": 1}  # optional for CSV imports
)

print("Connection engine created successfully!")

```

```python
# Create df and paste the file path for the CSV

import pandas as pd

df = pd.read_csv(r"C:\Users\Usuario\Desktop\Analyst Builder\Data analytics project Restaurant Health Inspection Analysis - NYC\Files\DOHMH_New_York_City_Restaurant_Inspection_Results_20250919.csv")

print(df.head())
```

```python
#First do the following:
#1 Create a database in MySQL: 
#CREATE DATABASE restaurant_health_inspection;
#2 Verify the database exists
#SHOW DATABASES;

# Export to sql
df.to_sql(
    name="inspection_results",   # Table name
    con=engine,
    if_exists="replace",              # or "append" to add without deleting existing
    index=False,                      # Avoid exporting pandas index
    chunksize=1000                    # Handle large datasets in batches
)
```
